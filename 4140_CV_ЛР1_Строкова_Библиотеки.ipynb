> Колоризация с помощью библиотеки matplotlib


```python
%%time
# импортируем библиотеку
import matplotlib as mpl
# задаем пользовательскую шкалу
cmap = (mpl.colors.ListedColormap(['#A52A2A', '#FFA500', '#FFFF00', '#00FF00']))
# колоризируем
img = plt.imread(r'Dog.jpeg')
plt.imshow(img, cmap)
plt.axis('off')
plt.show()

print('#A52A2A - brown')
print('#FFA500 - orange')
print('#FFFF00 - yellow')
print('#00FF00 - green')
```

```python
# посмотрим на оригинал изображения
img = plt.imread(r'Dog.jpeg')
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.show()
```

```python
from PIL import Image, ImageDraw
image1 = Image.new('RGB', (90, 90), 'white')
draw = ImageDraw.Draw(image1)
draw.rectangle((0, 0, 90, 90), '#A52A2A')
plt.axis('off')
plt.imshow(image1)
```

```python
image2 = Image.new('RGB', (90, 90), 'white')
draw = ImageDraw.Draw(image2)
draw.rectangle((0, 0, 90, 90), '#FFA500')
plt.axis('off')
plt.imshow(image2)
```
