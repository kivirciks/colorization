```python
# импорт библиотек
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# обратимся к изображению
img = plt.imread(r'Dog.jpeg')
# выведем изображение на экран
plt.imshow(img, cmap='gray')
# уберем "рамку" matplotlib
plt.axis('off')
plt.show()
```

```python
# выведем тип данных
type(img)
```

```python
# создадим массив, который задает цветовую шкалу
array = np.random.randn(25, 25)
# выведем пустые квадраты, которые на следующих шагах заполнятся значениями цветовых шкал
fig, axs = plt.subplots(1, 2, figsize=(10,3))
# зададим черно-белую цветовую шкалу
p1 = axs[0].imshow(array, cmap='gist_gray')
fig.colorbar(p1, ax=axs[0])
# зададим цветовую шкалу в оттенках зелено-синего
p2 = axs[1].imshow(array, cmap='winter')
fig.colorbar(p2, ax=axs[1])
```

```python
# проверим, что цветовая карта gist_gray соответствует исходному изображению
plt.imshow(img, cmap='gist_gray')
plt.show()
```
