> Колоризация с помощью циклов

```python
from PIL import Image, ImageDraw
```

```python
# A52A2A - brown - rgb(165, 42, 42)
# FFA500 - orange - rgb(255, 165, 0)
# FFFF00 - yellow - rgb(255, 255, 0)
# 00FF00 - green - rgb(0, 255, 0)

image = Image.open('Dog.jpeg').convert('L').convert('RGB')
draw = ImageDraw.Draw(image) 
width, height = image.size
pix = image.load()

def colorize(image):
    for i in range(width):
        for j in range(height):
            r = pix[i, j][0]
            if 0 <= r <= 64:
                color = (165, 42, 42)
            elif 65 <= r <= 129:
                color = (255, 165, 0)
            elif 130 <= r <= 195:
                color = (255, 255, 0)
            else:
                color = (0, 255, 0)
            draw.point((i, j), color)
    image.save("PIL.jpeg", "JPEG")
```

```python
%%time
colorize(image)
image
```
