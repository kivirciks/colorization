> Колоризация с помощью библиотеки PIL





```python
# импорт библиотек
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
```

```python
# обратимся к изображению
img = plt.imread(r'Dog.jpeg')
# выведем изображение на экран
plt.imshow(img, cmap='gray')
# уберем "рамку" matplotlib
plt.axis('off')
plt.show()
```

```python
# создадим массив, который задает цветовую шкалу
rgb_array = [[255.000, 56.026 + (255 - 56.026) * i / 400, 255 * i / 400] for i in range(400)]
rgb_array += [[255 - 255 * i / 600, 255 - 255 * i / 600, 255] for i in range(600)]
img = np.array(rgb_array, dtype=int).reshape((1, len(rgb_array), 3))
# выведем цветовую шкалу на экран
plt.imshow(img, extent=[0, 25, 0, 25], aspect='auto')
plt.show()
```

```python
%%time
# обратимся к изображению, convert('L') - вывод в черно-белых оттенках
img = Image.open("Dog.jpeg").convert("L")
# выполним колоризацию
img1 = ImageOps.colorize(img, black ="blue", white ="red")
# выведем изображение на экран
plt.imshow(img1)
```
